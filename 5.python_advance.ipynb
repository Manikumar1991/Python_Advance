```python
#Q2.What is the concept of delegation?
'''
Delegation is an object oriented technique (also called a design pattern). Let's say you have an object x and want to change the behaviour of just one of its 
methods. You can create a new class that provides a new implementation of the method you're interested in changing and delegates all other methods to the 
corresponding method of x.
'''
```

```python
#Q5.What is the purpose of pseudoprivate attributes?
'''
private means the attributes are only available for the members of the class not for the outside of the class.
'''
class P:
   def __init__(self, name, alias):
      self.name = name       # public
      self.__alias = alias   # private

   def who(self):
      print('name  : ', self.name)
      print('alias : ', self.__alias)
```
