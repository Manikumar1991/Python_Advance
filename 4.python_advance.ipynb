```python
#Q2.In what contexts do the two operator overloading methods manage printing?
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high
 
    def __iter__(self):
        return self
 
    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        else:
            self.current += 1
            return self.current - 1
 
 
for num in Counter(5, 15):
    print(num)
```

```python
#Q3. In a class, how do you intercept slice operations?
'''
The __getitem__ method is used for accessing list items, array elements, dictionary entries etc. slice is a constructor in Python that creates slice object to 
represent set of indices that the range(start, stop, step) specifies. __getitem__ method can be implement in a class, and the behavior of slicing can be 
defined inside it.

Syntax:
__getitem__(slice(start, stop, step))
```

```python
sliced ='abcde'.__getitem__(slice(0, 3, 1)) 
print(sliced)
```
